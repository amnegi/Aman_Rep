```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Example: Boston = pd.read_csv('path_to_boston_dataset.csv')

# Define the features to plot against MEDV
column_sels = ['LSTAT', 'INDUS', 'NX', 'PTRATIO', 'RM', 'TAX', 'DIS', 'AGE']

# Set a Seaborn style for better aesthetics
sns.set_style('whitegrid')

# Define a list of colors for each scatter plot
colors = ['blue', 'green', 'red', 'purple', 'orange', 'brown', 'pink', 'cyan']

# Create scatter plots for each feature against MEDV
fig, axs = plt.subplots(ncols=4, nrows=(len(column_sels) // 4) + 1, figsize=(20, 15))
axs = axs.flatten()

# Loop through each feature and create a scatter plot with different colors
for i, col in enumerate(column_sels):
    sns.scatterplot(x=Boston[col], y=Boston['MEDV'], ax=axs[i], color=colors[i % len(colors)], alpha=0.6)
    axs[i].set_title(f'Scatter Plot of {col} vs MEDV', fontsize=12)
    axs[i].set_xlabel(col)
    axs[i].set_ylabel('MEDV')

# Hide any empty subplots (if the number of features is not a multiple of 4)
for j in range(i + 1, len(axs)):
    axs[j].set_visible(False)

# Adjust layout for better appearance
plt.tight_layout()
plt.show()
```

```python
# Applyling Log Transformation to remove skewness
```

```python
# Display the first few rows to see the original values
print("Original crim and lstat values:\n", Boston[['CRIM', 'LSTAT']].head())

# Apply log transformation to CRIM and LSTAT to reduce skewness
# Adding a small constant (e.g., 1) to avoid log(0) issues
Boston['crim_log'] = np.log(Boston['CRIM'] + 1)
Boston['lstat_log'] = np.log(Boston['LSTAT'] + 1)

# Display the transformed values
print("\nLog-transformed crim and lstat values:\n", Boston[['crim_log', 'lstat_log']].head())
```

```python
# To check if the skewness is removed after using LOG Transformation in the above code


import matplotlib.pyplot as plt
import seaborn as sns

# Plot histograms for original and transformed values
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Original CRIM
sns.histplot(Boston['CRIM'], bins=30, kde=True, ax=axes[0, 0], color='blue')
axes[0, 0].set_title('Original CRIM Distribution')

# Log-transformed CRIM
sns.histplot(Boston['crim_log'], bins=30, kde=True, ax=axes[0, 1], color='green')
axes[0, 1].set_title('Log-Transformed CRIM Distribution')

# Original LSTAT
sns.histplot(Boston['LSTAT'], bins=30, kde=True, ax=axes[1, 0], color='purple')
axes[1, 0].set_title('Original LSTAT Distribution')

# Log-transformed LSTAT
sns.histplot(Boston['lstat_log'], bins=30, kde=True, ax=axes[1, 1], color='orange')
axes[1, 1].set_title('Log-Transformed LSTAT Distribution')

# Adjust layout
plt.tight_layout()
plt.show()
```
