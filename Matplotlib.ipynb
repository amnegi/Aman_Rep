## OOPS Style :

```python
#Line Plots: to understand the trends and patterns in data
import numpy as np #hypothetical array data
import matplotlib.pyplot as plt

z = np.linspace(0, 2, 100) # no. of samples to generate

#Step 1: Declaring the figure and the axis
# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  # Create a figure and an axes.

#Step 2: start feeding data; .plot method generates line plot
ax.plot(z, z, label='linear')  # Plot some data on the axes. #plot 1

ax.plot(z, z**2, label='quadratic')  # Plot more data on the axes... #plot 2

ax.plot(z, z**3, label='cubic')  # ... and some more.(labels are the name/legends)

#step 3: Naming your x and y axis
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.

ax.set_title("Simple Plot")  # Add a title to the axes.

ax.legend()  # Add a legend (important for the above labels to be pulled)

plt.show()
```

## PLT Style :

```python
z = np.linspace(0, 2, 100)#Return evenly spaced numbers over a specified interval
#fig, ax = plt.subplots() is not required as no separate declaration of figure and axis is required
#ax.plot

plt.plot(z, z, label='linear')  # Plot some data on the (implicit) axes.
plt.plot(z, z**2, label='quadratic')  # etc.
plt.plot(z, z**3, label='cubic')

plt.xlabel('x label')#set_xlabel
plt.ylabel('y label')#set_ylabel

plt.title("Simple Plot")#set_title

plt.legend()
plt.show()
```

```python
# x-axis values
x = [5, 2, 9, 4, 7]

# Y-axis values
y = [10, 5, 8, 4, 2]

# Function to line plot
plt.plot(x, y, marker="*", color="r", linestyle="-.", linewidth = 3, ms=20) # ms = marker size

# function to show the plot
plt.show()
```

```python
# x-axis values
x = [5, 2, 9, 4, 7]

# Y-axis values
y = [10, 5, 8, 4, 2]

# Function to plot
plt.plot(x,y, marker="*", color="k", linestyle="-.", ms = 20, mfc = 'r', mec = 'k')
# mfc = marker fill color, mec = marker edge color

# function to show the plot
plt.show()
```

```python
# x values
x = [5, 2, 9, 4, 7]

# Y values
y = [10, 5, 8, 4, 2]

# Z values
z = [15, 8, 7, 5, 6]

```

```python
# Function to plot
plt.plot(x, y, marker="*", color="r", label="first plot")#st line plot

plt.plot(y, z, marker="p", color="b", linestyle="-.", label="second plot") #2nd line plot

plt.plot(x, z, marker="+", color="g", linestyle = "--", label="third plot") #3rd line plot

plt.xlabel('Height')
plt.ylabel('age')

plt.title("Simple Plot")
plt.legend()

# function to show the plot
plt.show()
```

```python
#assign a background style
plt.style.use('bmh')

fig, ax = plt.subplots(3, 1,sharex=True)

ax[0].plot(x, y, marker="v", color="r", linestyle="--")#first row

ax[1].plot(y, z, marker="o", color="b", linestyle="-.")#second row

ax[2].plot(x, z, marker="+", color="g", linestyle="-")#third row

ax[0].set_ylabel("nikhil")
ax[1].set_ylabel("garima")
ax[2].set_ylabel("harika")

ax[1].set_xlabel("hello")
ax[2].set_xlabel("Simple Plot")
ax[0].set_title("Grid Plots")

# function to show the plot
plt.show()
```

```python
#applicable only if you are working with a single column or a single row
fig, (ax1, ax2, ax3) = plt.subplots(3, 1,sharex=True)
#ax1 is first row
#ax2 is second row/grid
#ax3 is third row/grid

ax1.plot(x, y, marker="v", color="r", linestyle="--")

ax2.plot(y, z, marker="o", color="b", linestyle="-.")

ax3.plot(x, z, marker="+", color="g", linestyle="-")

ax1.set_ylabel("nikhil")
ax2.set_ylabel("garima")
ax3.set_ylabel("harika")

ax3.set_xlabel("Simple Plot")

# function to show the plot
plt.show()
```

```python
fig, ax = plt.subplots()

Movie_type = ['Comedy','Action','Romance','Drama','Scifi']

Responses = [4,5,6,1,4]

Responses_std = statistics.stdev(Responses)

ax.bar(Movie_type, Responses,yerr=Responses_std)

ax.set_ylabel("Survey Responses")

plt.show()
```

```python
#plotting multiple data

Movie_type = ['Comedy','Action','Romance','Drama','Scifi']

Men_responses = [4,5,6,1,4]

Women_responses = [8, 2, 7, 2, 1]

import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.bar(Movie_type, Men_responses, color = 'b', width = 0.50, label="Men Response")

ax.bar(Movie_type, Women_responses, color = 'g', width = 0.25, label="Women Response")

ax.set_ylabel("Survey Responses")

ax.legend()

ax.set_xticklabels(Movie_type, rotation=45) # rotation helps in label overlaping (x-axis)

plt.show()
```

```python
#Stacked bar chart
import matplotlib.pyplot as plt
Movie_type = ['Comedy','Action','Romance','Drama','Scifi']

Men_responses = [4,5,6,1,4]

Women_responses = [8, 2, 7, 2, 1]


fig, ax = plt.subplots()

ax.bar(Movie_type, Men_responses, color = 'b')

ax.bar(Movie_type, Women_responses, bottom = Men_responses, color = 'r')


ax.set_ylabel("Survey Responses")
ax.set_xticklabels(Movie_type, rotation=45) # rotation helps in label overlaping (x-axis)

ax.legend(labels=['Men', 'Women'])

plt.show()
```

```python
np.random.seed(10) # the random values generated will remain same everywhere, same reproducibility
collectn_1 = np.random.normal(100, 10, 200) # normally distributed (100 mean, SD = 10 ,200 values)
collectn_2 = np.random.normal(80, 30, 200)
collectn_3 = np.random.normal(90, 20, 200)
collectn_4 = np.random.normal(70, 25, 200)
#It takes three arguments, mean and standard deviation of the normal distribution, and the number of values desired.

## combine these different collections into a list
data_to_plot = [collectn_1, collectn_2, collectn_3, collectn_4]

fig, ax = plt.subplots()

# Create the boxplot
ax.boxplot(data_to_plot)

plt.show()
```
