# Basic Data Visualization - Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Piece_Dim.csv")
```

```python
df.head()
```

```python
plt.hist(df['Length'])
# 10 bins by default
```

```python
plt.hist(df['Length'], bins =15)
plt.show()
# Histogram is used to summarize discrete or continuous data (measured on an interval scale) like population or historical data.
```

```python
plt.hist(df['Length'], bins=15,color='Red')
plt.xlabel("Length in mm")
plt.ylabel("Count")
plt.title("Histogram of piece length")
plt.show()
```

## Box and Whisker Plot

```python
# Used for summary statistics such as median, quartiles and inter-quartile range, the distribution of data points across a selected measure.
plt.boxplot(df['Length'])
plt.show()
```

```python
op1 = df[df['Operator']=='Op-1']
op2 = df[df['Operator']=='Op-2']
```

```python
op2.head()
```

```python
plt.boxplot([op1['Length'], op2['Length']])
plt.show()
```

## Line Plot

```python
plt.plot(df['Length'])
plt.show()
```

```python
plt.plot(df['Length'])
mean_1 = df['Length'].mean()
sd_1 = df['Length'].std()
plt.axhline(y=mean_1, color ='red', linestyle='-')
plt.axhline(y=(mean_1 +3*sd_1), color ='red', linestyle='--')
plt.axhline(y=(mean_1 -3*sd_1), color ='red', linestyle='--')
plt.show()
```

## Scatter Plot

```python
plt.scatter(df['Length'], df['Width'])
```

## Bar Plot

```python
op = ['Operator 1','Operator 2','Operator 3']
pieces = [50,35,70]
plt.bar(op,pieces)
plt.show()
```

```python
# Pass values through a list in the two parameters inside plt.bar(1,2)
low = df[df['Length']<95]['Length'].count()
mid = df[(df['Length']>=95) & (df['Length']<=105)]['Length'].count()
high = df[df['Length']>105]['Length'].count()

size = [low,mid,high]
group = ['<95', 'Between 95 and 105', '>105']
plt.bar(group,size)
plt.show()
```
