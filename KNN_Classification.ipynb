#### Predicting flower species

```python
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
iris = datasets.load_iris()
type(iris)
```

```python
iris.keys()
```

```python
iris.data
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
x = iris.data # input variables
y = iris.target
```

```python
#split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=123)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=8)
#assume k = 8

knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
#predict the labels on the test data
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix


print(confusion_matrix(y_test, y_pred))

print(classification_report(y_test, y_pred))
```

-Support is No. of observations in test data

-In matrix each column sum represents Predicted values 

-Rows are Actual values

-Diagonals (True Positives) in matrix represents No. of correct predictions for each label 

-Recall(Hit Ratio) = Individual level accuracy (%), that all actual values were correctly predicted (quantity, row wise)

-Precision = Proportion (%) of predictions made which are actually correct in each label (quality, column wise)

-f1-score = harmonic mean of Precision or Recall

```python
y_pred_train = knn.predict(x_train)
#predict the labels on the test data
```

```python
print(confusion_matrix(y_train, y_pred_train))

print(classification_report(y_train, y_pred_train))
```

**Tune and Optimize KNN in python using scikit-learn**

```python

from sklearn.model_selection import GridSearchCV

parameters = {"n_neighbors": range(1, 50)} # range given end is not included (till 49)

gridsearch = GridSearchCV(KNeighborsClassifier(), parameters, cv=5)

gridsearch.fit(x_train, y_train)

# fitting on trained data 
```

```python
gridsearch.best_params_
```

```python
# now use this grid search model to further use on test data, no need to repeat the process
y_pred = gridsearch.predict(x_test)
#n_neighbors=3
```

```python
print(confusion_matrix(y_test, y_pred))

print(classification_report(y_test, y_pred))
```

```python
# Checking overfitting on train data
y_predtrain = gridsearch.predict(x_train)
#n_neighbors=3
print(classification_report(y_train, y_predtrain))
```

97% accuracy with uniform weightage model

```python
# Tuning model for prefferential weightage
parameters={'n_neighbors': range(1, 50),
                         'weights': ['uniform', 'distance']}
GD = GridSearchCV(KNeighborsClassifier(),parameters,cv=5)
# default options --> uniform (all points in the neighborhood are weighted equally)
# another option --> distance (weights closer neighbors more heavily than farther neighbors)
```

```python
GD.fit(x_train,y_train)
```

```python
# according to this we are getting better at 6 n neighbors above was 3 for equal weightage
y_pred = GD.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
# To check overfitting 
y_predtrain = GD.predict(x_train)
print(classification_report(y_train, y_predtrain))
```

Now as it is overfitted, we wont consider prefferential weightage.

### KNN with PCA


```python
#standariztion process: using standard scaler
#scikit learn prime package: machine learning: sklearn
from sklearn.preprocessing import StandardScaler

#instantiate the standard scaler: create an instance
scaler = StandardScaler(copy=True, with_mean=True, with_std=True)

#fittng the scaler on data
#scaler.fit(df)#algo will apply all the required maths; mean and stdev of each variable

#samples_scaled = scaler.transform(df)
#result of the scaler is always in array form

samples_scaled = scaler.fit_transform(x)
print(samples_scaled)
print()
print(x)
```

```python
# Import PCA
from sklearn.decomposition import PCA

# Create PCA instance: model
model = PCA() #all pcs will be used

# Apply the fit_transform method of model to grains: pca_features
x = model.fit_transform(samples_scaled)

#principal components or PCA features
x.shape
```

```python
# Now seeing how much variation explained by PCs
import numpy as np

print(np.round(model.explained_variance_ratio_ * 100, 1))
```

The PC1 and PC2 explains total of 95.9 % variation and we use these two again in above model.

But we dont know what we will drop or use, as the IVs are now unidentitfied as PCs.

```python
# Import PCA
from sklearn.decomposition import PCA

# Create PCA instance: model
model = PCA(n_components = 2) # PC1 and PC2 will be used

# Apply the fit_transform method of model to grains: pca_features
x = model.fit_transform(samples_scaled)

#principal components or PCA features
x.shape
```

```python
#split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=123)
#x here represents input features post PCA
```

```python
#implement classification using KNN
from sklearn.neighbors import KNeighborsClassifier

#Gridsearchcv to tune the value of k along with cross validation
from sklearn.model_selection import GridSearchCV

#declare the dictionary of the hyperparameter to be tuned
parameters = {"n_neighbors": range(1, 50), "weights": ['uniform', 'distance']}#1 to 49

gridsearch = GridSearchCV(KNeighborsClassifier(), parameters, cv=5)
#simple cross validation, not repeated

gridsearch.fit(x_train, y_train)
#, "weights": ['uniform', 'distance']
```

```python
#using gridsearch model to make predictions on the test data
y_pred = gridsearch.predict(x_test)

print(classification_report(y_test, y_pred))
```

The accuracy is lower but we check for overfit :

```python
y_predtrain = gridsearch.predict(x_train)

print(confusion_matrix(y_train, y_predtrain))

print(classification_report(y_train, y_predtrain))
```

**This model is overfitted, and we now will use uniform wt model with grid search (97%)**
