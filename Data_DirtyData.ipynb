### Replace empty values

```python
import pandas as pd
data = pd.read_csv("data.csv")
data
```

```python
data.isna().sum()
```

```python
data["Calories"].mean()
```

#### fillna replaces null values with specified values, but if we want original data we have to call again

```python
#update in script as per pandas 3
data.fillna({"Calories": data["Calories"].mean()},inplace=True)
```

```python
data.isna().sum()
```

```python
data = pd.read_csv("data.csv")
#replaced missing values with median
data.fillna({"Calories": data["Calories"].median()},inplace=True)
```

```python
data = pd.read_csv("data.csv")
data["Calories"].mode()
```

```python
# mode always come this 0,1,2 index position so later we called with [0] to replace missing values with mode
data = pd.read_csv("data.csv")
data.fillna({"Calories": data["Calories"].mode()[0]},inplace=True)
```
