# Index return data analysis of 4 Indices

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv("index.csv")
```

```python
# Saving the original data and creating a copy for analysis
df_comp = raw_data.copy()
```

```python
df_comp.spx.plot(figsize = (20,5), title = "S&P500 Prices")
plt.xlabel ("Day", fontsize = 14)
plt.ylabel("Index Value", fontsize = 14)
plt.show()
```

```python
df_comp.spx.plot(figsize = (20,5), title = "S&P500 Prices")
df_comp.ftse.plot(figsize = (20,5), title = "FTSE100 Prices")
plt.title("S&P v FTSE")
plt.show()
```

```python
df_comp.dax.plot(figsize = (20,5), title = "S&P500 Prices")
df_comp.nikkei.plot(figsize = (20,5), title = "FTSE100 Prices")
plt.title("DAX v NIKKEI")
plt.show()
```

```python
# All indices on same figure
plt.figure(figsize=(20,10))

# Plot index with proper labeling and color
plt.plot(df_comp.spx, label ='S&P500', color='blue', linewidth=2)
plt.plot(df_comp.ftse, label ='FTSE100', color='green', linewidth=2)
plt.plot(df_comp.dax, label ='DAX', color='red', linewidth=2)
plt.plot(df_comp.nikkei, label ='NIKKEI', color='orange', linewidth=2)

# Adding titles and lables

plt.title("Comparison of S&P500, FTSE100, DAX and NIKKEI Indices", fontsize=16)
plt.xlabel("Day",fontsize=14)
plt.ylabel("Index Value",fontsize=14)

# ADding a legend to distinguish the lines
plt.legend(loc='upper left',fontsize=12)

# SAving the image in IDE home
plt.savefig('indices_comparison.png', bbox_inches='tight') 

# DIsplaying the plot
plt.show()
```
