```python
Titanic['age'].fillna(Titanic['age'].mean(), inplace=True)
Titanic
```

```python
Titanic.dropna(subset=['embarked'], inplace=True)
Titanic
```

```python
Titanic.drop_duplicates(inplace=True)
Titanic
```

```python
Titanic['age'] = Titanic['age'].round(2)
Titanic['fare'] = Titanic['fare'].round(2)

# Display the first 5 rows to verify rounding
print(Titanic[['age', 'fare']].head())
```

```python
# ELIMINATE IRRELEVANT INFORMATION
# Drop the specified columns from the Titanic DataFrame
Titanic = Titanic.drop(columns=["name", "ticket", "cabin","boat","body","home.dest" ])

# Display the first few rows to verify the changes
print(Titanic.head())
```

```python
#This script creates a 1x3 grid of bar charts, each visualizing the frequency distribution of a categorical variable (Pclass, Sex, and Embarked) from the Titanic dataset.
#The countplot function from seaborn is used to display the count of occurrences for each category in the selected variables.
#The figure size and layout are adjusted for better readability.
```

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic_df = sns.load_dataset('titanic')

# Set up the figure and axes for multiple bar charts
fig, axes = plt.subplots(1, 3, figsize=(18, 5))  # 1 row, 3 columns of bar charts

# Plot the frequency of 'Pclass'
sns.countplot(ax=axes[0], x='pclass', data=titanic_df, palette='pastel')
axes[0].set_title('Passenger Class Frequency')
axes[0].set_xlabel('Passenger Class')
axes[0].set_ylabel('Count')

# Plot the frequency of 'Sex'
sns.countplot(ax=axes[1], x='sex', data=titanic_df, palette='cool')
axes[1].set_title('Sex Frequency')
axes[1].set_xlabel('Sex')
axes[1].set_ylabel('Count')

# Plot the frequency of 'Embarked'
sns.countplot(ax=axes[2], x='embarked', data=titanic_df, palette='muted')
axes[2].set_title('Embarkation Port Frequency')
axes[2].set_xlabel('Embarked')
axes[2].set_ylabel('Count')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
# 1c. ANALYSIS OF MULTIVARIATE VARIABLES
# After analyzing the characteristics one by one, it is time to analyze them in relation
#to the predictor and to themselves, in order to draw clearer conclusions about their relationships
# 1d NUMERICAL-NUMERICAL ANALYSIS - SURVIVED - (FARE, AGE)
# Scatter Plot with Hue (for Grouping)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a scatter plot with 'Age' on the x-axis and 'Fare' on the y-axis
sns.scatterplot(data=Titanic, x='age', y='fare', hue='survived')
plt.title('Scatter Plot of Fare vs Age Grouped by Survival Status')
plt.show()
```

```python
# 1e- CATEGORICAL-CATEGORICAL ANALYSIS
# SURVIVED - (SEX, PCLASS, EMBARKED, SIBSP, PARCH)
```

```python
fig, axis = plt.subplots(2, 3, figsize = (15, 7))

sns.countplot(ax = axis[0, 0], data = Titanic, x = "sex", hue = "survived")
sns.countplot(ax = axis[0, 1], data = Titanic, x = "pclass", hue = "survived").set(ylabel = None)
sns.countplot(ax = axis[0, 2], data = Titanic, x = "embarked", hue = "survived").set(ylabel = None)
sns.countplot(ax = axis[1, 0], data = Titanic, x = "sibsp", hue = "survived")
sns.countplot(ax = axis[1, 1], data = Titanic, x = "parch", hue = "survived").set(ylabel = None)

plt.tight_layout()
fig.delaxes(axis[1, 2])

plt.show()
```

```python
# The following conclusions can be drawn from the above graph:
#With higher proportion, women survived as opposed to men. This is because women had priority over men in the evacuation plans.
#People who traveled alone had more problems surviving than those who traveled accompanied.
#Those who traveled in a better class on the Titanic had a higher chance of survival.
```

```python
# COMBINATIONS OF CLASS WITH VARIOUS PREDICTOR
# The class and sex of the passenger versus their survival analysis is presented below.
# "Pclass" as the hue parameter means that different classes will be represented by different colors in the bars.
```

```python
fig, axis = plt.subplots(figsize = (10, 5), ncols = 2)

sns.barplot(ax = axis[0], data = Titanic, x = "sex", y = "survived", hue = "pclass")
sns.barplot(ax = axis[1], data = Titanic, x = "embarked", y = "survived", hue = "pclass").set(ylabel = None)

plt.tight_layout()

plt.show()
```

```python
# From these analyses, it is clear that, women had a higher chance of survival regardless of the port of embarkation
#and the class in which they traveled.
#Furthermore, on average, people who traveled in higher classes survived longer than those who did not.
```
