# PCA

**PCA follows the fit/transform pattern**

The fit method learns how to shift and how to rotate the samples, but doesn't actually change them.

The transform method, on the other hand, applies the transformation that fit learned.

In particular, the transform method can be applied to new, unseen samples

```python
import pandas as pd
df=pd.read_csv('seed.csv')
df
```

```python
#standarDiztion process: using standard scaler
#scikit learn prime package: machine learning: sklearn
from sklearn.preprocessing import StandardScaler

#instantiate the standard scaler: create an instance
scaler = StandardScaler(copy=True, with_mean=True, with_std=True)

#fittng the scaler on data
#scaler.fit(df)#algo will apply all the required maths; mean and stdev of each variable

#samples_scaled = scaler.transform(df)
#result of the scaler is always in array form

samples_scaled = scaler.fit_transform(df)
print(samples_scaled)
print()
print(df)
```

```python
# Import PCA
from sklearn.decomposition import PCA

# Create PCA instance: model
model = PCA() #all pcs will be used

# Apply the fit_transform method of model to grains: pca_features
pca_features = model.fit_transform(samples_scaled)

#principal components or PCA features
pca_features
```

**When PCA is implemented, it leads to decorrelation of data**

```python
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Assign 0th column of pca_features: xs
pc1 = pca_features[:,0]

# Assign 1st column of pca_features: ys
pc2 = pca_features[:,1]

# Scatter plot xs vs ys
plt.scatter(pc1, pc2)
plt.axis('equal')
plt.show()

# Calculate the Pearson correlation of xs and ys
correlation, pvalue = pearsonr(pc1, pc2)

# Display the correlation
print(correlation)
```

**TO demostrate how PCA is super intelligent to detect the correlations and patterns in high dimensional data, we are considering a seeds.csv**

This file consists of 7 columns where each column or feature relates to some of the physical characteristic of the underlying seed variety.



```python
import pandas as pd
grainsnew=pd.read_csv('seeds.csv')
grainsnew
```

```python
#standarDiztion process: using standard scaler
#scikit learn prime package: machine learning: sklearn
from sklearn.preprocessing import StandardScaler

#instantiate the standard scaler: create an instance
scaler = StandardScaler(copy=True, with_mean=True, with_std=True)

#fittng the scaler on data
#scaler.fit(df)#algo will apply all the required maths; mean and stdev of each variable

#samples_scaled = scaler.transform(df)
#result of the scaler is always in array form

grains = scaler.fit_transform(grainsnew)
```

```python
#running PCA on grains data
#pca will analyse data in high dimension
from sklearn.decomposition import PCA

pca = PCA()#instance of PCA created

projected = pca.fit_transform(grains)#7inputcolumns
#pca features

print(grains.shape)#original dataframe
print()
print(projected.shape)#PCA features/components
```

```python
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Assign 0th column of pca_features: xs
pc1 = projected[:,0]

# Assign 1st column of pca_features: ys
pc3 = projected[:,2]

# Scatter plot xs vs ys
plt.scatter(pc1, pc3)
plt.axis('equal')
plt.show()

# Calculate the Pearson correlation of xs and ys
correlation, pvalue = pearsonr(pc1, pc3)

# Display the correlation
print(correlation)
```

When PCA converts each input column into a principal component, then it sorts the PCs into the sorted i.e. descending order. This sort is performed in terms of variation explained i,e, which component is explaining maximum variance or variation in the input data. AT any stage, target variable is not considered.

TO understand the significance of a variable in terms of its contribution to the target variable.

The draback is, I dont know which PC is representing which input column

No predefined method to do the reverse enginnering

We dont need to do that. The idea of PCA is to do dimension reduction and it is recommended in those algos which use the distance metric logic.

```python
import numpy as np
np.round(pca.explained_variance_ratio_ * 100,1)
```

```python
#extract the amount of explained variation by each principal component
import numpy as np

print(np.round(pca.explained_variance_ratio_ * 100, 1))
```

```python
pcanew = PCA(n_components=3)#reinstantiating PCA with significant components only

grainsnew = pcanew.fit_transform(grains)#original data is grains consisting of 7 features

grainsnew.shape#only top 2 components have been produced
```

```python
#########################
#
# Draw a scree plot and a PCA plot
#
#########################
import numpy as np
import matplotlib.pyplot as plt

#extract the variance values of each PC: explained_variance_ratio_ * 100

#The following code constructs the Scree plot
per_var = np.round(pca.explained_variance_ratio_* 100, decimals=1)#result in array form
#percentage of variation that each principal component accounts for

labels = ['PC' + str(x) for x in range(1, 8)] #PC1 range(1,8): 1,2,3,4,5,6,7
#creating labels : one label per principal component

plt.bar(x=range(1,8), height=per_var, tick_label=labels)

plt.ylabel('Percentage of Explained Variance')
plt.xlabel('Principal Component')
plt.title('Scree Plot')
plt.show()
```
