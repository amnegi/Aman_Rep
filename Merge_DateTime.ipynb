## Merging two dataframes
To merge we need at least one common column in both the tabel/dataframes

```python
player = ['Player1','Player2','Player3']
point = [8,9,6]
title = ['Game1','Game2','Game3']
df1 = pd.DataFrame({'Player':player, 'Points': point, 'Title': title})
df1#left table
```

```python
player = ['Player1', 'Player5', 'Player6']
power = ['Punch', 'kick', 'Elbow']
title = ['Game1', 'Game5','Game6']
df2 = pd.DataFrame({'Player': player, 'Power': power, 'Title': title})
df2 #right table
```

```python
df1.merge(df2)
```

 Specific to merge , other common column's multiples are made 

```python
df1.merge(df2, on='Player', how='inner')
```

```python
df1.merge(df2, on='Title', how='inner')
```

Only common elements are present other are Nan

```python
df1.merge(df2,how='left')
```

```python
df1.merge(df2,how='right')
```

### merging multiple dataframes

```python

player = ['Player1','Player2','Player3', 'Player4']
score = [88,92,63, 22]
df1 = pd.DataFrame({'Player':player, 'Scores': score})
print(df1)

player = ['Player1','Player2','Player4']
wicket = [0,1,5]
df2 = pd.DataFrame({'Player':player, 'Wickets': wicket})
print(df2)

player = ['Player4','Player2','Player3']
catches = [0, 2, 1]
df3 = pd.DataFrame({'Player':player, 'Catch': catches})
print(df3)

player = ['Player4','Player5','Player3']
fifty = [0, 2, 1]
df4 = pd.DataFrame({'Player':player, 'Fifty': fifty})
print(df4)
```

```python
from functools import reduce 
data_frames = [df1, df2, df3, df4] # creating list to use in reduce funtion, as reduce it takes two parameters

# a,b is df1,df2 then it's result becomes a and df3 becomes b , so merging done on a,b and so on.... 

df_merged = reduce(lambda  a,b: pd.merge(a, b ,on=['Player'],
                                            how='outer'), data_frames)

df_merged
```

### Date Time

```python
from datetime import datetime

now = datetime.now() # current date and time
print(now)
```

```python
# time stamp according to us

year = now.strftime("%Y")
print("year:", year)

month = now.strftime("%m")
print("month:", month)

day = now.strftime("%d")
print("day:", day)

time = now.strftime("%H:%M:%S")
print("time:", time)

date_time = now.strftime("%m/%d/%Y, %H:%M:%S")
print("date and time:",date_time)
```

```python
timestamp =1528797322
date_time = datetime.fromtimestamp(timestamp)
date_time
```

```python
# above is year, day, month, hours, minutes, seconds
# now making them in proper format

timestamp = 1528797322
date_time = datetime.fromtimestamp(timestamp)

print("Date time object:", date_time)

d = date_time.strftime("%m/%d/%Y, %H:%M:%S")
print("Output 2:", d)

d = date_time.strftime("%d %b, %Y")
print("Output 3:", d)

d = date_time.strftime("%d %B, %Y")
print("Output 4:", d)

d = date_time.strftime("%I%p")
print("Output 5:", d)
```

```python
# different types of timestamp available to us from different modifiers (eg:- %c)
```

```python
#locale based

timestamp = 1528797322
date_time = datetime.fromtimestamp(timestamp)

d = date_time.strftime("%c")
print("Output 1:", d)

d = date_time.strftime("%x")
print("Output 2:", d)

d = date_time.strftime("%X")
print("Output 3:", d)
```
