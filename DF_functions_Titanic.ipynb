## Segragating Columns
Removing or using columns as per need.

```python
# column name case sensitive calling 
age_sex = titanic[['age','sex']]
age_sex.head()
```

```python
# if one column is use then one bracket can be used but it will not be a df it will be a series
sex = titanic['sex']
sex
```

```python
# Now it is a dataframe using double brackets
sex = titanic[['sex']]
sex
```

```python
# writing  axis =1 is important to drop, and to make the drop permanent use inplace=True
titanic.drop(columns=["boat","body"], axis=1, inplace= True)
```

```python
titanic.head()
```

## Filtering of rows

```python
titanic[titanic["age"] > 35]
# if no connected to outside titanic or main df , it  will give boolean result
```

```python
# now the above rows's index are jumbled so we do:
titanic[titanic["age"] > 35].reset_index()
```

```python
# AND
titanic [ ( titanic["age"] > 35 ) & ( titanic["age"] < 60 ) ].reset_index()
```

```python
# OR
titanic[(titanic["pclass"] == 2) | (titanic["pclass"] == 3)].reset_index() 
```

```python
titanic[titanic["pclass"].isin([2,3])]
```

```python
titanic[titanic["age"].notna()]
```

```python
titanic.dropna(subset=["age"])
```

### Statistical Explorations on Dataframe

```python
titanic["age"].mean()
```

```python
titanic["age"].median()
```

```python
titanic[["age","fare"]].describe()
```

```python
titanic["sex"].describe()
```

```python
titanic.agg({"age": ["count","mean","median"], "fare": ["min","max","median"]})
```

```python
#What is the average age for male vs female titanic passengers?
#sex and age #mean
#aggregate statitics grouped the category: sex
#groupby

titanic[["sex","age"]].groupby("sex").mean()
```

```python
titanic.groupby("sex")["age"].mean()
```

```python
# What is the mean ticket fare price for each of the cabin class combination
```

```python
titanic.groupby("pclass")["fare"].mean()
```

```python
# What is the mean fare price for sex and cabin class combination?
# fare,pclass,sex : categorical
```

```python
titanic.groupby("sex")["fare"].mean()
```

```python
titanic.groupby(["sex","pclass"])["fare"].mean()
```

```python
# What is the average fare and age for each male and female class combination?
# sex, fare as well as age
```

```python
titanic.groupby(["sex","pclass"])[["fare","age"]].mean()
```

```python
titanic.groupby(["sex","pclass"])[["age","fare"]].describe()
```

Combining Groupby and agg funtions

```python
titanic.groupby(["sex","pclass"]).agg({"age": ["count","mean","median"], "fare": ["min","max","median"]})
```
