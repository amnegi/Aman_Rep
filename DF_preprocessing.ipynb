## Segragating Columns
Removing or using columns as per need.

```python
# column name case sensitive calling 
age_sex = titanic[['age','sex']]
age_sex.head()
```

```python
# if one column is use then one bracket can be used but it will not be a df it will be a series
sex = titanic['sex']
sex
```

```python
# Now it is a dataframe using double brackets
sex = titanic[['sex']]
sex
```

```python
# writing  axis =1 is important to drop, and to make the drop permanent use inplace=True
titanic.drop(columns=["boat","body"], axis=1, inplace= True)
```

## Filtering of rows

```python
titanic[titanic["age"] > 35]
# if no connected to outside titanic or main df , it  will give boolean result
```

```python
# now the above rows's index are jumbled so we do:
titanic[titanic["age"] > 35].reset_index()
```

```python
# AND
titanic [ ( titanic["age"] > 35 ) & ( titanic["age"] < 60 ) ].reset_index()
```

```python
# OR
titanic[(titanic["pclass"] == 2) | (titanic["pclass"] == 3)].reset_index() 
```

```python
titanic[titanic["pclass"].isin([2,3])]
```

```python
titanic[titanic["age"].notna()]
```
